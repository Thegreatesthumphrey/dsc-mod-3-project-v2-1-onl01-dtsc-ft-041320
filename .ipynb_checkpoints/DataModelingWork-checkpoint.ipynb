### Split for Training and Testing Data

```python
from sklearn.model_selection import train_test_split
drop1  = ['churn', 'state']
drop2 = ['churn', 'area_code','state', 'total_intl_charge', 'total_intl_calls']
drop3 = ['churn', 'area_code','state', 'total_intl_charge', 'total_intl_calls', 'number_vmail_messages']
X = df.drop(columns=drop1, axis=1)
y = df['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

### Balance Classifiers

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_sample(X_train, y_train) 
```

### Scale

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

scaled_data_train = scaler.fit_transform(X_train)
scaled_data_train_resampled = scaler.fit_transform(X_train_resampled)
scaled_data_test = scaler.transform(X_test)

scaled_df_train = pd.DataFrame(scaled_data_train, columns=X.columns)
scaled_df_test = pd.DataFrame(scaled_data_test, columns=X.columns)
```

## KNN Model

<p> *Note that RECALL is of high importance here, as we want to catch all possible customers that might leave even if ones that will not are included, so that we have a chance to change that outcome. </p>

### Fit/Predict

```python
from sklearn.neighbors import KNeighborsClassifier


classifier = KNeighborsClassifier(n_neighbors=17, weights='uniform')


classifier.fit(scaled_data_train_resampled, y_train_resampled)


test_preds = classifier.predict(scaled_data_test)
```

### Evaluate

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

def print_metrics(labels, preds):
    print("Precision Score: {}".format(precision_score(labels, preds)))
    print("Recall Score: {}".format(recall_score(labels, preds)))
    print("Accuracy Score: {}".format(accuracy_score(labels, preds)))
    print("F1 Score: {}".format(f1_score(labels, preds)))
    
print_metrics(y_test, test_preds)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, test_preds)
roc_auc = auc(false_positive_rate, true_positive_rate)
print('\nAUC is :{0}'.format(round(roc_auc, 2)))
```

```python
print('\nConfusion Matrix')
print('----------------')
pd.crosstab(y_test, test_preds, rownames=['True'], colnames=['Predicted'], margins=True)
```

### Findings
As I increased the number of n_neighbors, the recall and accuracy rate rose. It capped at 15 neighbors, and started lowering beyond that. I also test weights of 'uniform' and 'distance' at n=15 and 'uniform' was very slighlty higher in those areas. 
<br/><br/>
With n=15 and weights='uniform', the best output was: <br/>
Precision Score: 0.31645569620253167 <br/>
Recall Score: 0.8695652173913043<br/>
Accuracy Score: 0.7230215827338129<br/>

## DT Model

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
from IPython.display import Image, DisplayObject, display 
from pydotplus import graph_from_dot_data
```

```python
# Identify the optimal tree depth for given data
max_depths = np.linspace(1, 32, 32, endpoint=True)
train_results = []
test_results = []
for max_depth in max_depths:
   dt = DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, random_state=0)
   dt.fit(scaled_data_train_resampled, y_train_resampled)
   train_pred = dt.predict(scaled_data_train_resampled)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_train_resampled, train_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   # Add auc score to previous train results
   train_results.append(roc_auc)
   y_pred = dt.predict(scaled_data_test)
   false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred)
   roc_auc = auc(false_positive_rate, true_positive_rate)
   # Add auc score to previous test results
   test_results.append(roc_auc)

plt.figure(figsize=(12,6))
plt.plot(max_depths, train_results, 'b', label='Train AUC')
plt.plot(max_depths, test_results, 'r', label='Test AUC')
plt.ylabel('AUC score')
plt.xlabel('Tree depth')
plt.legend()
plt.show()
```

### Fit/Predict

```python
#Using a max_depth of 9
classifier = DecisionTreeClassifier(random_state=0, criterion='entropy', max_depth=9)

classifier.fit(scaled_data_train_resampled, y_train_resampled)

dt_y_preds = classifier.predict(scaled_data_test)
dt_y_train_preds = classifier.predict(scaled_data_train_resampled)
```

### Evaluate

```python
def print_metrics(labels, preds):
    print("Precision Score: {}".format(precision_score(labels, preds)))
    print("Recall Score: {}".format(recall_score(labels, preds)))
    print("Accuracy Score: {}".format(accuracy_score(labels, preds)))
    print("F1 Score: {}".format(f1_score(labels, preds)))
    
print_metrics(y_test, dt_y_preds)
print('Accuracy of Training Data: ', accuracy_score(y_train_resampled, dt_y_train_preds))
```

```python
from sklearn.metrics import accuracy_score, roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, dt_y_preds)
roc_auc = auc(false_positive_rate, true_positive_rate)
print('\nAUC is :{0}'.format(round(roc_auc, 2)))
```
