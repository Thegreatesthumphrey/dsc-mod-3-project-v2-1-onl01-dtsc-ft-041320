### Scale

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

scaled_data_train = scaler.fit_transform(X_train)
scaled_data_test = scaler.transform(X_test)

scaled_df_train = pd.DataFrame(scaled_data_train, columns=X.columns)
scaled_df_test = pd.DataFrame(scaled_data_test, columns=X.columns)
```

### Lasso for Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error

for feat in scaled_df_train.columns:
    scaled_data_tr = scaled_df_train[scaled_df_train[feat]]
    scaled_data_tes =scaled_df_test[scaled_df_test[feat]]
    scaled_data_tr = scaled_data_tr.array()
    lasso = Lasso()
    lasso.fit(scaled_data_tr, y_train)

    print('Training r^2:', lasso.score(scaled_data_tr, y_train))
    print('Test r^2:', lasso.score(scaled_data_tes, y_test))
    print('Training MSE:', mean_squared_error(y_train, lasso.predict(scaled_data_tr)))
    print('Test MSE:', mean_squared_error(y_test, lasso.predict(scaled_data_tes)))
```

## KNN Model

### Fit/Predict

```python
from sklearn.neighbors import KNeighborsClassifier


classifier = KNeighborsClassifier(n_neighbors=7, weights='distance')


classifier.fit(scaled_data_train, y_train)


test_preds = classifier.predict(scaled_data_test)
```

### Evaluate

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

def print_metrics(labels, preds):
    print("Precision Score: {}".format(precision_score(labels, preds)))
    print("Recall Score: {}".format(recall_score(labels, preds)))
    print("Accuracy Score: {}".format(accuracy_score(labels, preds)))
    print("F1 Score: {}".format(f1_score(labels, preds)))
    
print_metrics(y_test, test_preds)
```

```python
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, test_preds)
roc_auc = auc(false_positive_rate, true_positive_rate)
print('\nAUC is :{0}'.format(round(roc_auc, 2)))
print(false_positive_rate, true_positive_rate, thresholds)
```

### Improve

```python
def find_best_k(X_train, y_train, X_Test, y_test, min_k=1, max_k=25):
    best_k = 0
    best_score = 0.0
    for k in range(min_k, max_k+1, 2):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        preds = knn.predict(X_test)
        f1 = f1_score(y_test, preds)
        if f1 > best_score:
            best_k = k
            best_score = f1
    
    print("Best Value for k: {}".format(best_k))
    print("F1-Score: {}".format(best_score))
    
find_best_k(scaled_data_train, y_train, scaled_data_test, y_test)
```
